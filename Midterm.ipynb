```python
# Load the dataset
import pandas as pd
import os


# Check what files are available
#print("Available files:")
#for file in os.listdir('.'):
#    if file.endswith('.csv'):
#        print(f"  - {file}")
news_data_path = '/content/BBC News Train.csv'

# Load the main dataset (adjust filename as needed)
# Common filenames: train.csv, bbc-text.csv, or similar
news_data = pd.read_csv(news_data_path)  # Adjust filename based on what you see
news_data.describe()
#print(f"Dataset shape: {news_data.shape}")
#print(f"Columns: {news_data.columns.tolist()}")
#print(news_data.head())


#print(f"Categories: {news_data['category'].unique()}")  # Adjust column name as needed
```

```python
# Data preparation template
import pandas as pd
import numpy as np

# 1. Examine your dataset
print("Dataset Info:")
print(f"Shape: {news_data.shape}")
print(f"Columns: {news_data.columns.tolist()}")
print("\nFirst few rows:")
print(news_data.head())

# 2. Identify text and category columns
# Adjust these column names based on your dataset
text_column = 'Text'  # or 'description', 'content', 'headline', etc.
category_column = 'Category'  # or 'label', 'class', etc.

# 3. Check for missing values
print(f"\nMissing values:")
print(news_data.isnull().sum())

# 4. Remove rows with missing text or categories
news_data_clean = news_data.dropna(subset=[text_column, category_column])

# 5. Check category distribution
print(f"\nCategory distribution:")
print(news_data_clean[category_column].value_counts())

# 6. Sample if dataset is too large (keep under 2000 for Colab)
if len(news_data_clean) > 2000:
    news_data_final = news_data_clean.sample(n=2000, random_state=42)
    print(f"\nSampled dataset to {len(news_data_final)} rows")
else:
    news_data_final = news_data_clean

# 7. Rename columns for consistency
news_data_final = news_data_final.rename(columns={
    text_column: 'content',
    category_column: 'category'
})

# 8. Save prepared dataset
news_data_final.to_csv('newsbot_dataset.csv', index=False)
print("\n✅ Dataset prepared and saved as 'newsbot_dataset.csv'")
```
